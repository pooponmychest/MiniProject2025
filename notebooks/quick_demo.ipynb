```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Folder where process_data.py saves processed CSVs
processed_dir = "../data/processed"

# List processed CSVs
files = [f for f in os.listdir(processed_dir) if f.endswith("_processed.csv")]
print("Available files:")
for i, f in enumerate(files, 1):
    print(f"{i}. {f}")
```

```python
# Pick a file by number
choice = int(input("Select file number: ")) - 1
file_name = files[choice]

# Load the file
df = pd.read_csv(os.path.join(processed_dir, file_name), parse_dates=["date"], index_col="date")
print(f"Loaded {file_name}")
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["close"], label="Close")
plt.plot(df.index, df["sma_10"], label="SMA 10")
plt.plot(df.index, df["sma_50"], label="SMA 50")
plt.title("Stock Price with Moving Averages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["vol_20"], label="20-day Volatility")
plt.title("Annualized Volatility")
plt.legend()
plt.show()
```
